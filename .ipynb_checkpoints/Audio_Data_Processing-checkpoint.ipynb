```python
import pandas as pd 
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

from glob import glob

import librosa
import librosa.display
import IPython.display as ipd

from itertools import cycle
```

```python
sns.set_theme(style = "white", palette = None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
# readin in the audio files using glob
audio_files = glob('archive/Piano_Notes/*.wav') # go through all the file names and return the wav file for each
```

```python
# listen to one of these files
ipd.Audio(audio_files[0])
```

```python
# read in the file using load
y, sr = librosa.load(audio_files[2]) # y = raw data, sr = sample rate
```

```python
y.shape
```

```python
sr
```

```python
# plotting the wave to get an idea of what it looks like
# easier to plot as pd, series to give x and y values and plot from pandas to plot it
pd.Series(y).plot(figsize = (10,5), lw = 1, title = 'Raw Audio Example', color = color_pal[0])
plt.show()
```

```python
# remove the silence within the audio file
y_trim, _ = librosa.effects.trim(y, top_db = 10)
```

```python
y_trim.shape
```

```python
pd.Series(y_trim).plot(figsize = (10,5), lw = 1, title = 'TrimData', color = color_pal[1])
plt.show()
```

# Calculate the Chormagram
## Represent the distribution of the pitch energy levels

```python
chroma = librosa.feature.chroma_stft(y = y, sr = sr)
```

```python
chroma.shape  # number of pich classes with time frame
```

### extracting the chorma

```python
# get all the chroma values until a specific time frame
chroma_at_frame = chroma[:,66]
```

```python
prominant_pitch = np.argmax(chroma_at_frame)
```

```python
pitch_classes = ["C","C#","D","D#","E","F","F#","G","G#","A","A#","B"]
```

```python
pitch_name = pitch_classes[prominant_pitch]
pitch_name
```

### extracting the chorma for trim

```python
chroma_trim = librosa.feature.chroma_stft(y = y_trim, sr = sr)
```

```python
chroma_trim.shape
```

```python
chroma_at_frame = chroma[:,21]
pitch = np.argmax(chroma_at_frame)
pitch_name = pitch_classes[pitch]
```

```python
pitch_name
```

```python
stft_trim = np.abs(librosa.stft(y = y_trim))
```
